# CityBikes

Send a request to CityBikes for the city of your choice.

```python
import pandas as pd
import os
import requests
import json
```

```python
#set the keys
url = "http://api.citybik.es/v2/networks/divvy"
# Create dictionary for headers
headers = {"Accept": "application/json"}
# Add key with our API KEY
params = {"fields": "location.latitude,location.longitude,greenbike"}
result = requests.get(url,params=params)

```

```python
data = result.json()
data

```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes).

Put your parsed results into a DataFrame.

```python
n = len(data['network']['stations'])
columns = ['name', 'renting ', 'returning','free_bikes','total_bikes','longitude','latitude']
df = pd.DataFrame(columns=columns, index=None)
i = 0
while i < n:
    renting = data['network']['stations'][i]['extra']['renting']
    returning = data['network']['stations'][i]['extra']['returning']
    latitude = data['network']['stations'][i]['latitude']
    longitude = data['network']['stations'][i]['longitude']
    free_bikes = data['network']['stations'][i]['free_bikes']
    name = data['network']['stations'][i]['name']
    total_bikes = renting + returning + free_bikes
    data1 = {'name': name , 'renting ': renting , 'returning': returning , 'free_bikes': free_bikes ,'total_bikes' : total_bikes,'longitude': longitude , 'latitude':latitude }
    df = df.append(data1, ignore_index=True)
    i = i + 1

    
```
